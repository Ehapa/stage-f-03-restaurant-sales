```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#%matplotlib.inline
```

```python
#read in train dataset
data =pd.read_csv('../data/train.csv')
```

```python
#drop 'Id' column
data.drop(['Id'], axis=1, inplace = True)
```

```python
#convert 'Open Date' column 
data['Open Date'] = pd.to_datetime(data['Open Date']).dt.strftime('%Y%m%d')
print(data['Open Date'])
```

```python
data.info()
```

```python
#dealing with categirical variables
from sklearn.preprocessing import LabelEncoder

col= ('Open Date', 'City', 'City Group', 'Type')
for x in col:
    le= LabelEncoder()
    data[x] = le.fit_transform(data[x].values)
data.dtypes
```

```python
data.head()
```

```python
#scale data
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
X_std = sc.fit_transform(data)
```

```python
data.head()
```

```python
#compute covariance matrix
mean_vec = np.mean(X_std, axis=0)
cov_mat = (X_std - mean_vec).T.dot((X_std - mean_vec)) / (X_std.shape[0]-1)
print('Covariance matrix \n%s' %cov_mat)
```

```python
#perform eigendecompositon on covariance matrix
cov_mat = np.cov(X_std.T)
eig_vals, eig_vecs = np.linalg.eig(cov_mat)

print('Eigenvectors \n%s' %eig_vecs)
print('\nEigenvalues \n%s' %eig_vals)
```

```python
#visually confirm that the list is correctly sorted by decreasing eigenvalues
eig_pairs = [(np.abs(eig_vals[i]), eig_vecs[:, i]) for i in range(len(eig_vals))]
print('Eigenvalues in descending order:')
for i in eig_pairs:
    print(i[0])
```

```python
#import and apply Principal Component Analysis
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principalComponents_data = pca.fit_transform(data)
print(pca.explained_variance_ratio_)
```

```python
principalComponents_data
```

```python
#explained variance
pca = PCA().fit(X_std)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cummulative explained variance')
plt.show()
```

```python
#create a DataFrame that will have the principal component values for all 137 samples.
principal_Df = pd.DataFrame(data = principalComponents_data
             , columns=['principal component 1', 'principal component 2'])
```

```python
principal_Df.tail()
```

```python
#plot the visualization along the principal component - 1 and principal component - 2 axis
import seaborn as sns
plt.figure()
plt.figure(figsize=(10,10))
plt.xticks(fontsize=12)
plt.yticks(fontsize=14)
plt.xlabel('Principal Component - 1',fontsize=20)
plt.ylabel('Principal Component - 2',fontsize=20)
plt.title("Principal Component Analysis of Restaurant Revenue Prediction",fontsize=20)


y = data['revenue']
sns.scatterplot(principal_Df['principal component 1'] , principal_Df['principal component 2'], hue= y, s=100)
```
